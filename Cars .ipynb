```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# opening cardekho.com
url = "https://www.cardekho.com/"
driver.get(url)
time.sleep(4)

driver.maximize_window()
time.sleep(2)
```

```python
#Click on used car menu 
Used_car_click=driver.find_element_by_xpath('//span[@class="used-cars-india"]')
Used_car_click.click()
time.sleep(5)
```

```python
#select the city 
city_click= driver.find_element_by_xpath('//a[@title="Used cars in  Kolkata"]')
city_click.click()
time.sleep(2)
```

```python
# scrape all Cars urls for Jaipur
Car_urls = []
url=driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]/a')
for i in url:
        Car_urls.append(i.get_attribute("href"))
        
```

```python
len(Car_urls)
```

```python
# Make Empty List of the required data for Kolkata
Brand= []

Make_Year= []
Fuel=[]
KMs_Driven=[]
Engine_Displacement=[]
No_Of_Owner=[]
Transmission=[]
Mileage=[]
Max_Power=[]
Torque=[]

Seats=[]
Color=[]
Gear_Box=[]
Drive_Type=[]
Steering_Type=[]
Front_Brake_Type=[]
Rear_Brake_Type=[]

Acceleration=[]
Engine_Type=[] 
Price= []
```

```python
for url in Car_urls[576:]:
    driver.get(url)
    time.sleep(2)
    print('Sraping ', url)
    
    
    
    # Extracting Price 
    try:
        price = driver.find_element_by_xpath('//div[@class="priceSection"]/span[2]')      
        Price.append(price.text)
    except:
        price1= driver.find_element_by_xpath("//span[@class='amount']")    
        Price.append(price1.text)

    time.sleep(1)
    
    # Extracting Brand from xpath
    try:
        brand = driver.find_element_by_xpath('//div[@class="gsc_col-xs-12"]/h1')      
        Brand.append(brand.text)
    except:
        brand1= driver.find_element_by_xpath("//div[@class='vdp-head']/h1[2]")    
        Brand.append(brand1.text)

    time.sleep(1)
    
    # Extracting make year
    try:
        year = driver.find_element_by_xpath('(//div[@class="iconDetail" or @class="fontweight500 value"])[1]')      
        Make_Year.append(year.text)
    except NoSuchElementException:
        Make_Year.append('-')

    time.sleep(1)
    
    
   # Extracting Fuel Type 
    try:
        fuel = driver.find_element_by_xpath('(//div[@class="fontweight500 value"])[3]')      
        Fuel.append(fuel.text)
    except:
        fuel1= driver.find_element_by_xpath('(//div[@class="iconDetail"])[5]')    
        Fuel.append(fuel1.text)

    time.sleep(1)
    
    # Extracting KM Driven  
    try:
        km_driven = driver.find_element_by_xpath('(//div[@class="fontweight500 value"])[4]')      
        KMs_Driven.append(km_driven.text)
    except:
        km_driven1= driver.find_element_by_xpath('(//div[@class="iconDetail"])[3]')    
        KMs_Driven.append(km_driven1.text)

    time.sleep(1)
    
    # Extracting Engine_Displacement  
    try:
        engine_displacement = driver.find_element_by_xpath('(//div[@class="fontweight500 value"])[5]')      
        Engine_Displacement.append(engine_displacement.text)
    except:
        engine_displacement1= driver.find_element_by_xpath('(//div[@class="largeSpec"])[2]')    
        Engine_Displacement.append(engine_displacement1.text)

    time.sleep(1)
    
    # Extracting No_Of_Owner
    try:
        no_Of_owner = driver.find_element_by_xpath('(//div[@class="fontweight500 value"])[6]')      
        No_Of_Owner.append(no_Of_owner.text)
    except:
        no_Of_owner1= driver.find_element_by_xpath('(//div[@class="iconDetail"])[4]')    
        No_Of_Owner.append(no_Of_owner1.text)

    time.sleep(1)
    
    # Extracting Transmission
    try:
        transmission = driver.find_element_by_xpath('(//div[@class="fontweight500 value"])[8]')      
        Transmission.append(transmission.text)
    except:
        transmission1= driver.find_element_by_xpath('(//div[@class="iconDetail"])[6]')    
        Transmission.append(transmission1.text)

    time.sleep(1)
    
    # Open full specification 
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight*0.35);")
        time.sleep(1)
        Spec = driver.find_element_by_xpath("//div[text()='View All Specifications']")      
        Spec.click()
    except:
        full_spec=driver.find_element_by_xpath('(//div[@class="brdrBtm"]/a)[2]')
        full_spec.click()

    time.sleep(2)
    
        # Extracting Mileage
    try:
        mileage = driver.find_element_by_xpath('(//div[@class="fontweight500 value"])[10]')      
        Mileage.append(mileage.text)
    except:
        mileage1= driver.find_element_by_xpath('(//div[@class="largeSpec"])[1]')    
        Mileage.append(mileage1.text)

    time.sleep(1)
    
    # Extracting Max_Power
    try:
        max_power = driver.find_element_by_xpath('(//div[@class="fontweight500 value"])[12]')      
        Max_Power.append(max_power.text)
    except:
        max_power1= driver.find_element_by_xpath('(//div[@class="largeSpec"])[3]')    
        Max_Power.append(max_power1.text)

    time.sleep(1)
    
    # Extracting Torque
    try:
        torque = driver.find_element_by_xpath('(//div[@class="fontweight500 value"])[13]')      
        Torque.append(torque.text)
    except:
        torque1= driver.find_element_by_xpath('(//div[@class="largeSpec"])[4]')    
        Torque.append(torque1.text)

    time.sleep(1)
    
    # Extracting Seats
    try:
        seats = driver.find_element_by_xpath('//div[text()="Seats"]/following-sibling:: div')      
        Seats.append(seats.text)
    except NoSuchElementException:
        Seats.append('-')

    time.sleep(1)
    
    # Extracting Color
    try:
        color = driver.find_element_by_xpath('//div[text()="Color"]/following-sibling:: div')      
        Color.append(color.text)
    except NoSuchElementException:
        Color.append('-')

    time.sleep(1)
    
        # Extracting Gear_Box
    try:
        gear_Box = driver.find_element_by_xpath('//div[text()="Gear Box"]/following-sibling:: div')      
        Gear_Box.append(gear_Box.text)
    except NoSuchElementException:
        Gear_Box.append('-')

    time.sleep(1)
    
        
    # Extracting Drive_Type
    try:
        drive_Type = driver.find_element_by_xpath('//div[text( )="Drive Type"]/following-sibling:: div')      
        Drive_Type.append(drive_Type.text)
    except NoSuchElementException:
        Drive_Type.append('-')

    time.sleep(1)
    
        # Extracting Steering_Type
    try:
        steering_Type = driver.find_element_by_xpath('//div[text( )="Steering Type"]/following-sibling:: div ')      
        Steering_Type.append(steering_Type.text)
    except NoSuchElementException:
        Steering_Type.append('-')

    time.sleep(1)
    
        # Extracting Front_Brake_Type
    try:
        front_Brake_Type = driver.find_element_by_xpath('//div[text( )="Front Brake Type"]/following-sibling:: div')      
        Front_Brake_Type.append(front_Brake_Type.text)
    except NoSuchElementException:
        Front_Brake_Type.append('-')

    time.sleep(1)
    
    # Extracting Front_Brake_Type
    try:
        rear_Brake_Type = driver.find_element_by_xpath('//div[text( )="Rear Brake Type"]/following-sibling:: div ')      
        Rear_Brake_Type.append(rear_Brake_Type.text)
    except NoSuchElementException:
        Rear_Brake_Type.append('-')

    time.sleep(2)
    
        # Extracting Front_Brake_Type
    try:
        acceleration = driver.find_element_by_xpath('//div[text( )="Acceleration"]/following-sibling:: div ')      
        Acceleration.append(acceleration.text)
    except NoSuchElementException:
        Acceleration.append('-')

    time.sleep(1)
    
        # Extracting Front_Brake_Type
    try:
        engine_Type = driver.find_element_by_xpath('//div[text( )="Engine Type"]/following-sibling:: div ')      
        Engine_Type.append(engine_Type.text)
    except NoSuchElementException:
        Engine_Type.append('-')

    time.sleep(1)
```

```python
# Making dataFrame for Kolkata cars
Kolkata=pd.DataFrame({})
Kolkata ['Brand']= Brand

Kolkata ['Make_Year']= Make_Year
Kolkata ['Fuel']=Fuel
Kolkata ['KMs_Driven']= KMs_Driven
Kolkata ['Engine_Displacement']= Engine_Displacement
Kolkata ['No_Of_Owner']= No_Of_Owner
Kolkata ['Transmission']= Transmission
Kolkata ['Mileage']= Mileage 

Kolkata ['Max_Power']= Max_Power 
Kolkata ['Torque']= Torque

Kolkata ['Seats']= Seats 
Kolkata ['Color']= Color 
Kolkata ['Gear_Box']= Gear_Box 
Kolkata ['Drive_Type']= Drive_Type 
Kolkata ['Steering_Type']= Steering_Type 
Kolkata ['Front_Brake_Type']= Front_Brake_Type 
Kolkata ['Rear_Brake_Type']= Rear_Brake_Type 

Kolkata ['Acceleration']= Acceleration 
Kolkata ['Engine_Type']= Engine_Type 
Kolkata ['Price']= Price 
```
